##Array Properties

```python
arr = np.array([[1,2,3],
                 [4,5,6]])
print("Shape :", arr.shape)         # return the shape of the matrix or vector or tensor whatever you can say (n X m)
print("Dimension :", arr.ndim)      # return the dimension of the array
print("Size :", arr.size)           # return total no. of element in the array
print("DataType :", arr.dtype)
```

##Array Reshaping

```python
arr = np.arange(25)
print("\n Original Array :",arr)

reshaped = arr.reshape(5,5)                     # convert the into (n x m) using reshape method
print("\n Reshaped Array :\n",reshaped)

flatten = reshaped.flatten()
print("\n Flatten Array :",flatten)             # convert the reshape array into original array ( return copy of reshaped)

raveled = reshaped.ravel()
print("\n Raveled Array :",raveled)             # return the view (original) instead of copy

#Transpose
transpose = reshaped.T
print("\n Transpose Array :\n",transpose)
```

```python
arr = np.arange(21)
print(arr)

#Slicing
print("\nBasic slicning :", arr[2:7])
print("\nWith step slicning :", arr[2:8:2])
print("\nNegatived slicning :", arr[-3])
```

##Sorting


```python
unsorted = np.array([2,3,1,65,3,43,6,1,4,2,64,2,5,2,3])
print("Sorted Array :", np.sort(unsorted))

unsorted_2d_arr = np.array([[4,2], [4,1], [2,3]])
print("Sorted 2D array column-wise", np.sort(unsorted_2d_arr, axis=0))
print("Sorted 2D array column-wise", np.sort(unsorted_2d_arr, axis=1))
```

##Filter

```python
number = np.arange(21)
even_numbers = number[number % 2 == 0]
print(even_numbers)


odd_numbers = number[number % 2 != 0]
print(odd_numbers)
```

##Filter By Mask

```python
mask = number > 5
print("Numbers greater than 5 :", number[mask])
mask
```

```python
where_result = np.where(number > 5)
print("Where :", number[where_result])
print(where_result)
```

```python
num = np.arange(11)
conditional_array = np.where(num > 5, num * 10, num)
print(conditional_array)


conditional_array = np.where(num > 5, "True", "False")
print(conditional_array)
```

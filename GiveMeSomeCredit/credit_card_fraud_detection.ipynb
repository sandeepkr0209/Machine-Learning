## Peform Simple imputer to fill the values with median

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
df['NumberOfDependents'] = imputer.fit_transform(df[['NumberOfDependents']])
df['MonthlyIncome'] = imputer.fit_transform(df[['MonthlyIncome']])
```

```python
X1 = df.drop('SeriousDlqin2yrs', axis=1)
y1 = df['SeriousDlqin2yrs']
```

```python
X1.isnull().sum()
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
X_train1.isnull().sum()
```

# performing the standard sacler

```python
from sklearn.preprocessing import StandardScaler
scaled = StandardScaler()
```

```python
X_train_scale = scaled.fit_transform(X_train1)
X_test_scale = scaled.transform(X_test1)
```

```python
X_train_scale.shape
```

# Training the Logistic Regression model using balanced as class_weights

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

model = LogisticRegression(class_weight='balanced', max_iter=500)
model.fit(X_train_scale, y_train1)

y_pred_prob = model.predict_proba(X_test1)[:, 1]
```

```python
threshold = 0.3
y_pred3 = (y_pred_prob > threshold).astype(int)
```

```python
precision_score(y_test1, y_pred3)
```

```python
recall_score(y_test1, y_pred3)
```

```python
r2_score(y_test1, y_pred3)
```

```python
print("ROC-AUC:", roc_auc_score(y_test, y_pred_prob))
print(confusion_matrix(y_test, y_pred3))
```

```python
for t in [0.1, 0.2, 0.3, 0.4]:
    y_pred = (y_pred_prob > t).astype(int)
    print(t, roc_auc_score(y_test, y_pred_prob))
```

# Now lets try SMOTE for imbalanced data and logistic regression

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

sm = SMOTE()
rus = RandomUnderSampler()

X_over, y_over = sm.fit_resample(X_train1, y_train1)
X_res, y_res = rus.fit_resample(X_over, y_over)
```

```python
model2 = LogisticRegression(max_iter=500)
model2.fit(X_res, y_res)
```

```python
y_pred_prob2 = model2.predict_proba(X_test1)[:, 1]

threshold = 0.3
y_pred4 = (y_pred_prob2 > threshold).astype(int)
```

```python
precision_score(y_test1, y_pred4)
```

```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data structure : restaurant_id, 2021,2022,2022,2023,2024,2025

sales_data = np.array([
    [1, 200000, 300000, 400000, 500000, 600000, 200000],
    [2, 300000, 100000, 900000, 120000, 400000, 1000000],
    [3, 400000, 500000, 450000, 500000, 650000, 2000000],
    [4, 100000, 400000, 440000, 501000, 600000, 5000000],
    [5, 300000, 100000, 900000, 120000, 400000, 1000000]
])
print(sales_data)

# df = pd.DataFrame(sales_data)
# df
```

```python
print("Sales Data Shape :", sales_data.shape)
```

```python
print("Sales Data Size :", sales_data.size)
```

```python
sd = sales_data[:, 1:]
print(sd)
```

```python
sum = np.sum(sd, axis=0) # column wise using axis = 0
for i in range(len(sd)):
  print(f"Yearly sales of all restaurant across each year {2020 + i + 1} :", sum[i])
```

```python
min_values = np.min(sd, axis=1)
for i in range(len(sd)):
  print(f"Min Sales of each restaurant {i+1} across all year :", min_values[i])
```

```python
avg = np.average(sd, axis=1)  # for average we can np.mean also it gives the same answer
for i in range(len(sd)):
  print(f"Avg sales of each restaurant {i+1} across all year : { avg[i]:.2f}")
```

```python
max_restaurant_indices_per_year = np.argmax(sd, axis=0)
years_start = 2021
for i in range(sd.shape[1]):
  restaurant_index = max_restaurant_indices_per_year[i]
  restaurant_id = restaurant_index + 1  # Assuming restaurant_id starts from 1
  year = years_start + i
  max_sales_value = sd[restaurant_index, i]
  print(f"In year {year}, Restaurant {restaurant_id} had the maximum sales of : {max_sales_value}")
```

```python
cumsum = np.cumsum(sd, axis=1) # column wise using axis = 0
print(cumsum)
```
